```
# Compute Global Correlation between % Iron Concentrate and % Silica Concentrate
global_corr = df[['% Iron Concentrate', '% Silica Concentrate']].corr().iloc[0, 1]
print(f"Global correlation: {global_corr:.3f}")
```

```
# Compute daily correlations
df['date_only'] = df['date'].dt.date
daily_corrs = (
    df.groupby('date_only')
    .apply(lambda g: g[['% Iron Concentrate', '% Silica Concentrate']].corr().iloc[0, 1])
    .reset_index(name='daily_corr')
)
```

```
# Flag days that deviate from the global pattern
threshold = 0.30  # change as needed
daily_corrs['deviation'] = (daily_corrs['daily_corr'] - global_corr).abs()
daily_corrs['flag'] = daily_corrs['deviation'] > threshold
```

```
# Plot visualization of daily correlation between % Iron and % Silica
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 5))
sns.lineplot(data=daily_corrs, x='date_only', y='daily_corr', marker='o',color='grey', markerfacecolor='grey')

plt.axhline(global_corr, color='red', linestyle='--', label='Global Correlation')
plt.title('Daily Correlation Between % Iron and % Silica')
plt.ylabel('Pearson Correlation')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```
# Plot visualiztion of daily correlation between % Iron and % Silica
# Define the threshold value for the flagged days
threshold = 0.30

plt.figure(figsize=(12, 5))

# Line for daily correlation
sns.lineplot(data=daily_corrs, x='date_only', y='daily_corr', marker='o', color='grey', markerfacecolor='grey')

# Global correlation baseline
plt.axhline(global_corr, color='black', linestyle='--', label='Global Correlation')

# Add threshold boundaries
plt.axhline(global_corr + threshold, color='red', linestyle='--', alpha=0.5, label='Flagged Threshold')
plt.axhline(global_corr - threshold, color='red', linestyle='--', alpha=0.5)

plt.title('Daily Correlation Between % Iron and % Silica')
plt.ylabel('Pearson Correlation')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```
# Show only flagged days
flagged_days = daily_corrs[daily_corrs['flag']]
print("Flagged days with significant correlation deviation:")
print(flagged_days)
```

```
# Fixing the column names in the code to match the actual column names in the DataFrame
def get_corr_pair(g, var1, var2):
    return g[[var1, var2]].corr().iloc[0, 1]

# Correcting the column names based on the DataFrame's actual column names
daily_corrs = df.groupby('date_only').apply(lambda g: pd.Series({
    'iron_vs_silica': get_corr_pair(g, '% Iron Concentrate', '% Silica Concentrate'),
    'iron_vs_ph': get_corr_pair(g, '% Iron Concentrate', 'Ore Pulp pH'),
    'silica_vs_ph': get_corr_pair(g, '% Silica Concentrate', 'Ore Pulp pH'),
})).reset_index()
```

```
# Plot Visualization of Daily Correlation Between Key Variables  
plt.figure(figsize=(14, 6))
sns.lineplot(data=daily_corrs, x='date_only', y='iron_vs_silica', label='Iron vs Silica', color ='grey')
sns.lineplot(data=daily_corrs, x='date_only', y='iron_vs_ph', label='Iron vs pH', color = 'plum')
sns.lineplot(data=daily_corrs, x='date_only', y='silica_vs_ph', label='Silica vs pH', color = 'pink')
plt.axhline(0, color='grey', linestyle='--', linewidth=1)
plt.ylabel('Pearson Correlation')
plt.xlabel('Date')
plt.title('Daily Pearson Correlation Between Key Variables')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt

# Define colors
color_iron_ph = '#6A0DAD'   # Deep purple
color_silica_ph = '#FF69B4' # Hot pink
color_iron_silica = 'grey'
color_global_corr = 'grey'

# Global correlation value (mean of Iron vs Silica, or replace with fixed value)
global_corr = daily_corrs['iron_vs_silica'].mean()

plt.figure(figsize=(14, 6))

# Plot each line and store handle
line1, = plt.plot(daily_corrs['date_only'], daily_corrs['iron_vs_ph'], label='Iron vs pH', color=color_iron_ph)
line2, = plt.plot(daily_corrs['date_only'], daily_corrs['silica_vs_ph'], label='Silica vs pH', color=color_silica_ph)
line3, = plt.plot(daily_corrs['date_only'], daily_corrs['iron_vs_silica'], label='Iron vs Silica', color=color_iron_silica)

# Global correlation line
line4 = plt.axhline(global_corr, color=color_global_corr, linestyle='--', label='Iron vs Silica Global Correlation')

# Horizontal reference at 0
plt.axhline(0, color='black', linestyle='--', linewidth=1)

# Custom legend in your desired order
plt.legend(handles=[line1, line2, line3, line4], loc='best')

# Final formatting
plt.ylabel('Pearson Correlation')
plt.xlabel('Date')
plt.title('Daily Pearson Correlation Between Key Variables')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
